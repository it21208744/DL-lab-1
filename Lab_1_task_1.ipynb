```python
x = np.linspace(-5, 5, 100)  # Create x-axis values
y = np.linspace(-5, 5, 100)  # Create y-axis values
X, Y = np.meshgrid(x, y)  # Create a grid of x and y values
Z = X**2 + Y**2  # Calculate Z values

fig = plt.figure(figsize=(8, 6))  # Adjust plot size
ax = fig.add_subplot(111, projection='3d')  # Create a 3D subplot
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)  # Plot the surface with colormap and transparency
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
